```python
import glob
import pandas as pd
import numpy as np
```

```python
states = pd.read_csv('states.csv')
# for Code 1995
states = states.set_index(states["id"].apply(lambda x: x.replace(" ", "").upper()))
```

```python
for i in ['11','12','13','14']:
    table = pd.DataFrame(index = list(states['id']) , columns = list(states['id']))
    table.index.name = "State"
    table = table.fillna(0)
    filepaths = glob.glob('state*flow'+ str(i) + '*.csv')
    for path in filepaths:    
    #     year = path[path.find("/")+1:path.find("/")+5]
        # out is -1/ in is 1
        IO = (-1, 1)[path[path.find("state")+5]=="i"]
#         State = path[path.rfind("/")+5:path.rfind("/")+7]
#         State = State.replace(" ", "").upper()
#     #     print State


        df = pd.read_csv(path)
        data = (df.ix[:, 0:5])
        data.columns = ["tStateId","stateId","stateShort", "stateName", "Number"]
        data = data[data.Number.notnull()]
        data = data[data.stateName.notnull()]
        data = data[4:]

        for s in data.iterrows():
            if s[1]["stateId"] == "96":
                continue
            if(int(s[1][0]) == int(s[1][1])):
                table.loc[s[1][0]][s[1][0]] = int(s[1][4])
                continue

            # fill in table
            try:
                tStateId = states.loc[s[1][0]]["id"]
        #              print tStateId
            except:
                continue
            
            try:
                table.loc[tStateId][s[1][1]] += int(s[1][4])*IO
            except:
                pass
            
    #     break
    for index, row in table.iterrows():
        for col, j in row.iteritems():
            if index != col:
                if table.loc[index][col] != -table.loc[col][index]:
                    print str(col) + "and" + str(index) + "year" + str(i) + "failure"
    print table
    break
    table.to_csv("table" + str(i) + ".csv")

```
